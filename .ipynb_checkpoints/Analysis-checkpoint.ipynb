### Pie Chart of Gender

- Visualize the distribution for the variable 'gender'.
- There are slighlty more females (52.5%) than males (47.5%). The dataset is balanced. 

```python
# calculate number of individuals per gender 
category_counts = df['Gender'].value_counts()
```

```python
# create pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Category Distribution')
plt.show()
```

### Average Age of individuals 

```python
mean_age = df['Age'].mean()
mean_age
```

```python
# Transform from float to int 
mean_age_int = int(mean_age)
mean_age_int
```

### Line Graph 
- Generate a line graph that represents the number of people from
 the top 5 nationalities in the dataset.

```python
# check unique values for Nationality 
nationality_counts = df['Nationality'].value_counts()
```

```python
# Create line graph with top 5 nationalities 
top_nationalities = nationality_counts.head(5)
```

```python
# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(top_nationalities.index, top_nationalities.values, marker='o')
plt.xlabel('Nationality')
plt.ylabel('Number of Individuals')
plt.title('Number of Individuals by Top Nationalities')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
